# Chart Gallery with jupyter_cc

This notebook demonstrates using `%cc` to create various charts and visualizations.
Each cell sends a natural-language prompt to Claude, which generates and inserts
the plotting code for you to review and run.

**Workflow:**
1. Run a `%cc` cell with your chart request
2. Claude creates a new code cell with the implementation
3. Review and execute the generated cell
4. Continue iterating with follow-up prompts
