# Pandas Workflow with jupyter_cc

This notebook demonstrates using `%cc` for iterative pandas data manipulation.
Each prompt builds on the previous one within the same conversation, so Claude
knows about variables and DataFrames created in earlier steps.

**Key patterns shown:**
- `%cc` for continuing a conversation (Claude remembers context)
- `%cc_new` for starting a fresh conversation

```python
%load_ext jupyter_cc_magic
```

## Step 1: Create Sample Data

Ask Claude to generate a realistic sales dataset.

```python
%cc Create a sample DataFrame with 100 rows of sales data (date, product, region, quantity, price)
```

## Step 2: Transform

Add computed columns and aggregate. Claude knows about the DataFrame from step 1.

```python
%cc Add a revenue column and group by product to show total revenue
```

## Step 3: Filter and Pivot

Narrow down to the top products and reshape the data.

```python
%cc Filter to top 5 products by revenue and create a pivot table by region
```

## Step 4: Fresh Conversation

Use `%cc_new` to start a completely new conversation. Claude will still see
session variables but won't have the previous conversation history.

```python
%cc_new Explain what transformations were applied to the data
```
