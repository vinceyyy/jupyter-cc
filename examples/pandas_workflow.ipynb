# Pandas Workflow with jupyter-cc

Data is created with regular Python. `%cc` handles transformations and analysis.

**Key patterns:**
- `%cc` continues a conversation (Claude remembers previous context)
- `%cc_new` starts a fresh conversation

```python
%load_ext jupyter_cc
```

## Create Sample Data

```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 100

df = pd.DataFrame(
    {
        "date": pd.date_range("2024-01-01", periods=n, freq="D"),
        "product": np.random.choice(
            ["Widget A", "Widget B", "Gadget X", "Gadget Y", "Gizmo Z", "Gizmo W", "Doohickey"], n
        ),
        "region": np.random.choice(["North", "South", "East", "West"], n),
        "quantity": np.random.randint(1, 50, n),
        "unit_price": np.round(np.random.uniform(5.0, 100.0, n), 2),
    }
)
df.head(10)
```

## Transform

Ask Claude to add computed columns and aggregate. Claude sees `df` in the variable tracker.

```python
%cc Add a revenue column (quantity * unit_price) to df, then show total revenue by product sorted descending
```

## Filter and Pivot

Continue the conversation — Claude knows about the revenue column from the previous step.

```python
%cc Filter df to the top 3 products by total revenue and create a pivot table showing revenue by region
```

## Fresh Conversation

`%cc_new` starts fresh — Claude sees session variables but not the previous conversation.

```python
%cc_new What columns does df have? Show dtypes and any missing values
```
